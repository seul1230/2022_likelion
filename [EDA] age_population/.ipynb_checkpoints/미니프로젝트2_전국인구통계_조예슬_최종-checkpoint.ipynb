#### **Ïó∞Î†πÍµ¨Î∂Ñ**
ÏïÑÎèô / Ï≤≠ÏÜåÎÖÑ / Ï≤≠ÎÖÑ / Ï§ëÏû•ÎÖÑ / ÎÖ∏ÎÖÑ

```python
# ages = [
# ÏïÑÎèô      '0~9ÏÑ∏',
# Ï≤≠ÏÜåÎÖÑ    '10~19ÏÑ∏',
# Ï≤≠ÎÖÑ      '20~29ÏÑ∏','30~39ÏÑ∏',
# Ï§ëÏû•ÎÖÑ    '40~49ÏÑ∏','50~59ÏÑ∏','60~69ÏÑ∏',
# ÎÖ∏ÎÖÑ      '70~79ÏÑ∏','80~89ÏÑ∏','90~99ÏÑ∏','100ÏÑ∏ Ïù¥ÏÉÅ']

ages = [['0~9ÏÑ∏'],
        ['10~19ÏÑ∏'],
        ['20~29ÏÑ∏','30~39ÏÑ∏'],
        ['40~49ÏÑ∏','50~59ÏÑ∏','60~69ÏÑ∏'],
        ['70~79ÏÑ∏','80~89ÏÑ∏','90~99ÏÑ∏','100ÏÑ∏ Ïù¥ÏÉÅ']]
cl = ['ÏïÑÎèô', 'Ï≤≠ÏÜåÎÖÑ', 'Ï≤≠ÎÖÑ', 'Ï§ëÏû•ÎÖÑ', 'ÎÖ∏ÎÖÑ']


# # ÏïÑÎèô 
# age_1 = ['0~9ÏÑ∏'] 
# # Ï≤≠ÏÜåÎÖÑ
# age_2 = ['10~19ÏÑ∏']
# # Ï≤≠ÎÖÑ
# age_3 = ['20~29ÏÑ∏','30~39ÏÑ∏']
# # Ï§ëÏû•ÎÖÑ
# age_4 = ['40~49ÏÑ∏','50~59ÏÑ∏','60~69ÏÑ∏']
# # ÎÖ∏ÎÖÑ
# age_5 = ['70~79ÏÑ∏','80~89ÏÑ∏','90~99ÏÑ∏','100ÏÑ∏ Ïù¥ÏÉÅ']
```

```python
def age_class(x):
    for i in range(len(ages)):
        if x in (ages[i]):
            return cl[i]
    return np.nan
age_class('70~79ÏÑ∏')
```

```python
df['Íµ¨Î∂Ñ'] = df['Ïó∞Î†πÎåÄ'].map(age_class)
```

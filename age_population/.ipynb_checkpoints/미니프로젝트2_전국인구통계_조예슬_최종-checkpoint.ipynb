#### **인구수 자료형 변환**

```python
df.info()
```

```python
df['인구수'] = df['인구수'].fillna('0')
df.isnull().sum()
```

```python
df['인구수'] = df['인구수'].str.replace(',','').astype(int)
```

```python
df.info()
```

#### **Column에 총인구수 추가**

```python
df2 = pd.DataFrame(df.groupby(['행정구역', '날짜'])['인구수'].sum()).reset_index()
```

```python
df2 = df2.rename(columns = {'인구수':'총인구수'})
df2
```

```python
df = pd.merge(left = df, right = df2, on = ['행정구역', '날짜'])
df
```

```python
df['인구비율 (%)'] = round(df['인구수'] / df['총인구수'],4) * 100
df 
```

#### **연령구분**
아동 / 청소년 / 청년 / 중장년 / 노년

```python
# ages = [
# 아동      '0~9세',
# 청소년    '10~19세',
# 청년      '20~29세','30~39세',
# 중장년    '40~49세','50~59세','60~69세',
# 노년      '70~79세','80~89세','90~99세','100세 이상']

ages = [['0~9세'],
        ['10~19세'],
        ['20~29세','30~39세'],
        ['40~49세','50~59세','60~69세'],
        ['70~79세','80~89세','90~99세','100세 이상']]
cl = ['아동', '청소년', '청년', '중장년', '노년']


# # 아동 
# age_1 = ['0~9세'] 
# # 청소년
# age_2 = ['10~19세']
# # 청년
# age_3 = ['20~29세','30~39세']
# # 중장년
# age_4 = ['40~49세','50~59세','60~69세']
# # 노년
# age_5 = ['70~79세','80~89세','90~99세','100세 이상']
```

```python
def age_class(x):
    for i in range(len(ages)):
        if x in (ages[i]):
            return cl[i]
    return np.nan
age_class('70~79세')
```

```python
df['구분'] = df['연령대'].map(age_class)
```
